```python
import numpy as np
import matplotlib.pyplot as plt
import mglearn

from sklearn.neighbors import KNeighborsClassifier

# Display plots inside Jupyter Notebook
%matplotlib inline

# Set random seed for reproducibility
np.random.seed(42)

# Number of samples
n_samples = 200

# Generate synthetic dataset with three features
X = np.random.randn(n_samples, 3)

# Create binary labels using a linear rule
y = (X[:, 0] + X[:, 1] + 0.5 * X[:, 2] > 0).astype(int)

print("X shape:", X.shape)

# Use only the first two features for visualization
X_2D = X[:, :2]

print("X_2D shape:", X_2D.shape)

# Scatter plot of the dataset
plt.figure(figsize=(6, 5))
mglearn.discrete_scatter(X_2D[:, 0], X_2D[:, 1], y)
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.legend(["Class 0", "Class 1"])
plt.title("Synthetic Dataset")
plt.show()

# Create plots for different K values
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for n_neighbors, ax in zip([1, 2, 5], axes):

    # Create and train the KNN model
    model = KNeighborsClassifier(n_neighbors=n_neighbors)
    model.fit(X_2D, y)

    # Plot decision boundary
    mglearn.plots.plot_2d_separator(
        model,
        X_2D,
        fill=True,
        eps=0.5,
        ax=ax,
        alpha=0.4
    )

    # Plot data points
    mglearn.discrete_scatter(
        X_2D[:, 0],
        X_2D[:, 1],
        y,
        ax=ax
    )

    # Labels and title
    ax.set_title(f"{n_neighbors} Neighbor(s)")
    ax.set_xlabel("Feature 0")
    ax.set_ylabel("Feature 1")

axes[0].legend(loc=4)

plt.tight_layout()
plt.show()
```
